```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from scipy.ndimage.filters import gaussian_filter1d
%matplotlib inline
```

```python
lst = pd.read_csv('./Boston/listings.csv')
lst.info()
```

```python
lst.drop(
    ['square_feet', 'has_availability','license','jurisdiction_names','neighbourhood_group_cleansed'], 
    axis=1, 
    inplace=True
)
```

```python
# convert price from string to float
lst['price'] = lst['price'].fillna('$0')
lst['price'] = lst['price'].apply(lambda x:float(x[1:].replace(',',''))) 
```

```python
lst.head()
```

## Area vs Price

```python
lst['city'].value_counts()
```

```python
lst['neighbourhood_cleansed'].value_counts() ## This look best
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.boxplot(
    data=lst, 
    x='neighbourhood_cleansed',
    y='price'
)
xlabel=ax.get_xticklabels()
ax.set_xticklabels(xlabel, rotation=90)
ax.set_ylim(0,550)
ax.set_title('Price vs Neighbourhood')
```

```python
price_neighbour = pd.DataFrame(
    lst[['longitude','latitude', 'price','neighbourhood_cleansed']].groupby('neighbourhood_cleansed').mean()
)

sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.scatterplot(
    data=lst,
    x='longitude',
    y='latitude', 
    size='price',
    color='orange',
)
ax = sns.scatterplot(
    data=price_neighbour,
    x='longitude',
    y='latitude', 
    size='price',
    color='blue'
)
ax.set_title('AirBnB Price Base on Area')
```

## Room Type vs Price

```python
room_type = lst[['room_type', 'price']].groupby('room_type').mean()
room_type
```

```python
ax=sns.countplot(
    data=lst,
    x='room_type'
)
ax.set(xlabel='Room Type', ylabel='Count', title='Total of Room Type')
```

```python
ax = sns.scatterplot(
    data=lst,
    x='longitude',
    y='latitude', 
    hue='property_type'
)
ax.set(title='Room Type Distribution')
```

```python
ax=sns.barplot(
    x='property_type',
    y='price',
    hue='room_type',
    data=lst,
    estimator=np.median
)
ax.set_xticklabels(labels=ax.get_xticklabels(),rotation=90)
ax.set_title('Price Base on Room Type and Property Type')
```

```python
lst['accommodates'].value_counts()
```

```python
acc_price = pd.DataFrame(
    lst[(lst['accommodates']>0) & (lst['price']>0)][['accommodates','price','id']].groupby('id').mean()
)
acc_price
```
